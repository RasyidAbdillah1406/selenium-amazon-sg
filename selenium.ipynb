```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from time import sleep
```

```python
# Open the browser
browser = webdriver.Chrome()

# Link to Amazon Singapore
browser.get("https://www.amazon.sg/")

# Maximize window browser
browser.maximize_window()

# Delay for 3 seconds
sleep(3)

# Refresh Browser
browser.refresh()

# Find input box 
input_search = browser.find_element(By.ID, 'twotabsearchtextbox')

# Search for items. In this case, We are looking for Crash Bandicoot
input_search.send_keys("Crash Bandicoot")

# Delay for 2 seconds
sleep(2)

# Click search button
input_search.send_keys(Keys.ENTER)


sleep(4)
# getVal = browser.find_element(By.XPATH, "//span[@class='a-size-base-plus a-color-base a-text-normal']")
# print(getVal)
```

```python
# Get the product details
for i in range(0,25):
    try:
        # Access the parent element 
        getVal = browser.find_element(By.XPATH, f"(//div[@class='a-section a-spacing-small puis-padding-left-small puis-padding-right-small'])[{i+1}]")
        
        # Access the title in the child element and take the value
        getTitle = getVal.find_element(By.XPATH, f"(.//span[@class='a-size-base-plus a-color-base a-text-normal'])")
        print(f"Title: {getTitle.text}")
        try:
            # Acess the type in the child element and take the value
            getType = getVal.find_element(By.XPATH, f"(.//a[@class='a-size-base a-link-normal s-underline-text s-underline-link-text s-link-style a-text-bold'])")
            print(f"Type: {getType.text}")
        except NoSuchElementException:
            # If the type is not provided, Then prompt the type is not stated             
            print(f"Type: Type is not available")
        try:
            # Access the price decimal in the child element and take the value             
            subPrice = getVal.find_element(By.XPATH, f"(.//span[@class='a-price-whole'])")
            # Access the price decimal in the child element and take the value             
            subPrice2 = getVal.find_element(By.XPATH, f"(.//span[@class='a-price-fraction'])")
            print("Price: S${0}.{1}".format(subPrice.text, subPrice2.text))
        except NoSuchElementException:
            # If the price is not stated, Then display the price is not stated             
            print("Price is not available")
        print("-----"*10, end="\n")
    # Display error if the element is not available
    except NoSuchElementException:
        print("Error")
```
